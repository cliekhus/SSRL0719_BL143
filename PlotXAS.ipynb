Here are the three dimer samples - with ACN, with Cl, and the oxidized Cl sample.  There are five major peaks.  The low energy peak for the chloride samples are from the Cl K-edge.  The main, large peak is from the Ru L-edge, the so called "B" peak.  There are two additional smaller peaks at 2845 and 2851 eV, which we attribute to "C" peaks, respective of the appropriate ligands.  Then there is a small shoulder 2838 eV, which is present for only the oxidized sample.  This is the "A" peak.

```python
#%matplotlib notebook
plt.figure(figsize=(12, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(XEnergy_RBCO3, XAS_RBCO3, marker='*', color = 'y', markerfacecolor = "None", markeredgecolor = 'y', label = "RuBpy2CO3")
plt.plot(XEnergy_RTB, XAS_RTB, marker = 'x', color = 'b', markerfacecolor = "None", markeredgecolor = 'b', label = "RuTBCN")
#plt.plot(XEnergy_R3D, XAS_R3D, marker = 'o', color = 'r', markerfacecolor = "None", markeredgecolor = 'r', label = "RuIIIDMAP")
#plt.plot(XEnergy_ACN, XAS_ACN, marker = 'D',  color = 'g', markerfacecolor = "None", markeredgecolor = 'g', label = "Ru Dimer w/ ACN")
#plt.plot(XEnergy_Cl, XAS_Cl, marker = '1', color = 'c', markerfacecolor = "None", markeredgecolor = 'c', label = "Ru Dimer w/ Cl")
#plt.plot(XEnergy_Clox, XAS_Clox, marker='s', color = 'm', markerfacecolor = "None", markeredgecolor = 'm', label = "Ru Dimer w/ Cl")
#plt.plot(XEnergy_RIID, XAS_RIID, marker='v', color = '0.5', markerfacecolor = "None", markeredgecolor = '0.5', label = "RuIIDMAP")
plt.plot([2845,2845], [-1,8], color = 'k', label = "CN C peak")
plt.plot([2851,2851], [-1,8], '--', color = 'k', label = "Py C peak")
plt.plot([2838,2838], [-1,8], ':', color = 'k', label = "RuIII A peak")
plt.xlim(2835, 2855)
plt.ylim(0,6.5)
plt.legend()
```

First step - identify the two separate C peaks.  Both of these monomers have pyridines, but only the monomer labeled "RuTBCN" has a CN bond.   Therefore we label the peak around 2845 eV as the CN C peak and the peak around 2851 eV as the pyridine C peak.

```python
#%matplotlib notebook
plt.figure(figsize=(12, 6), dpi= 80, facecolor='w', edgecolor='k')
#plt.plot(XEnergy_RTB, XAS_RTB, marker = 'x', color = 'b', markerfacecolor = "None", markeredgecolor = 'b', label = "RuTBCN")
#plt.plot(XEnergy_R3D, XAS_R3D, marker = 'o', color = 'r', markerfacecolor = "None", markeredgecolor = 'r', label = "RuIIIDMAP")
plt.plot(XEnergy_ACN, XAS_ACN, marker = 'D',  color = 'g', markerfacecolor = "None", markeredgecolor = 'g', label = "Ru Dimer w/ ACN")
plt.plot(XEnergy_Cl, XAS_Cl, marker = '1', color = 'c', markerfacecolor = "None", markeredgecolor = 'c', label = "Ru Dimer w/ Cl")
#plt.plot(XEnergy_Clox, XAS_Clox, marker='s', color = 'm', markerfacecolor = "None", markeredgecolor = 'm', label = "Ru Dimer w/ Cl")
#plt.plot(XEnergy_RBCO3, XAS_RBCO3, marker='*', color = 'y', markerfacecolor = "None", markeredgecolor = 'y', label = "RuBpy2CO3")
#plt.plot(XEnergy_RIID, XAS_RIID, marker='v', color = '0.5', markerfacecolor = "None", markeredgecolor = '0.5', label = "RuIIDMAP")
plt.plot([2845,2845], [-1,8], color = 'k', label = "CN C peak")
plt.plot([2851,2851], [-1,8], '--', color = 'k', label = "Py C peak")
plt.plot([2838,2838], [-1,8], ':', color = 'k', label = "RuIII A peak")
plt.xlim(2835, 2855)
plt.ylim(0,6)
plt.legend()
```

Comparing the two dimers, we see that the largest difference is due to a shift in the "B" peak to lower energies for the Cl ligand (more localized) compared to the ACN ligand (more delocalized).  This represents a shift in the eg energy.  

```python
#%matplotlib notebook
plt.figure(figsize=(12, 4), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(XEnergy_ACN, [x-y for x,y in zip(XAS_ACN,XAS_Cl)])
plt.xlim(2835, 2860)
plt.ylim(-0.6,0.6)
plt.title('ACN-Cl')
```

The shift is more obvious when taking the difference.  

```python
#%matplotlib notebook
plt.figure(figsize=(12, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(XEnergy_RIID, XAS_RIID, marker='v', color = '0.5', markerfacecolor = "None", markeredgecolor = '0.5', label = "RuIIDMAP")
#plt.plot(XEnergy_RTB, XAS_RTB, marker = 'x', color = 'b', markerfacecolor = "None", markeredgecolor = 'b', label = "RuTBCN")
plt.plot(XEnergy_R3D, XAS_R3D, marker = 'o', color = 'r', markerfacecolor = "None", markeredgecolor = 'r', label = "RuIIIDMAP")
#plt.plot(XEnergy_ACN, XAS_ACN, marker = 'D',  color = 'g', markerfacecolor = "None", markeredgecolor = 'g', label = "Ru Dimer w/ ACN")
plt.plot(XEnergy_Cl, XAS_Cl, marker = '1', color = 'c', markerfacecolor = "None", markeredgecolor = 'c', label = "Ru Dimer w/ Cl")
plt.plot(XEnergy_Clox, XAS_Clox, marker='s', color = 'm', markerfacecolor = "None", markeredgecolor = 'm', label = "Ru Dimer w/ Cl, oxidized")
#plt.plot(XEnergy_RBCO3, XAS_RBCO3, marker='*', color = 'y', markerfacecolor = "None", markeredgecolor = 'y', label = "RuBpy2CO3")
plt.plot([2845,2845], [-1,8], color = 'k', label = "CN C peak")
plt.plot([2851,2851], [-1,8], '--', color = 'k', label = "Py C peak")
plt.plot([2838,2838], [-1,8], ':', color = 'k', label = "RuIII A peak")
plt.xlim(2835, 2855)
plt.ylim(0,6.5)
plt.legend()
```

Now to look at the effect of oxidizing, we can compare the Ru(II) and R(III) monomers and the Ru(II)Cl dimer and the Ru(III)Cl dimer.  We see that oxidizing both samples shifts the peaks toward higher energies.  In addition, we can now clearly see the "A" peak around 2838 eV.  In the dimer, the "A" peak is not as strong as in the monomer.  

```python
#%matplotlib notebook
plt.figure(figsize=(12, 6), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(XEnergy_ACN, [x-y for x,y in zip(XAS_Clox,XAS_Cl)])
plt.plot([2838,2838], [-2,2])
plt.xlim(2835, 2860)
plt.ylim(-2,2)
plt.title('Clox-Cl')
```

A difference plot between the two Cl dimers, shows the obvious appearance of the "A" peak and a shift in the "B" energies.  The "C" peak is left unchanged.
