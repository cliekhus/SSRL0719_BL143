```python
import matplotlib.pyplot as plt
from averageScans import averageScans
import normAndOffset
```

```python
XEnergy_RTB, I0_RTB, FF1_RTB, FF2_RTB, FF3_RTB, FF4_RTB = averageScans('RuTpyBpyCN_10mM_090', 3, True)
```

```python
plt.plot(XEnergy_RTB, FF1_RTB)
plt.plot(XEnergy_RTB, FF2_RTB)
#plt.plot(XEnergy_RTB, FF3_RTB)
#plt.plot(XEnergy_RTB, FF4_RTB)
```

```python
XAS_RTB = normAndOffset.normAndOffset(I0_RTB, FF1_RTB, FF2_RTB, XEnergy_RTB, 2825, 2875)
plt.plot(XEnergy_RTB, XAS_RTB)
plt.xlabel('xray energy (eV)')
plt.title('XAS Ru_Tpy_Bpy (TFY)')
```

```python
XEnergy_R3D, I0_R3D, FF1_R3D, FF2_R3D, FF3_R3D, FF4_R3D = averageScans('RuIIIDMAP_5mM_099', 10, True)
```
